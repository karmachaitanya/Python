```python
mtcars=pd.read_csv("../input/mtcarsmtcars/mtcars.csv")
print(mtcars.to_string())
```

```python
mtcars.head()
```

```python
mtcars.columns
```

```python
#table 
pd.crosstab(mtcars.gear,mtcars.cyl)

#bar plot between 2 different categories 
pd.crosstab(mtcars.gear,mtcars.cyl).plot(kind="bar")
```

```python
mtcars["gear"].value_counts()
```

```python
mtcars.gear.value_counts().plot(kind="pie")
```

```python
plt.scatter(mtcars.mpg,mtcars.qsec)
```

```python
# Graphical Representation of data
#import matplotlib.pyplot as plt
# Histogram
plt.hist(mtcars['mpg'])
```

```python
#creates histogram with 5bins and colours filled init
plt.hist(mtcars['mpg'],facecolor ="peru",edgecolor ="blue",bins =5)
```

```python
#Boxplot
#help(plt.boxplot)
plt.boxplot(mtcars['mpg'],vert = True)
```

```python
#Boxplot
#help(plt.boxplot)
plt.boxplot(mtcars['mpg'],vert = False)
```

```python
#for horizontal
plt.boxplot(mtcars['mpg'],vert =False);
plt.ylabel("MPG");
plt.xlabel("Boxplot");
plt.title("Boxplot")
```

```python
#Violin Plot
#help(plt.violinplot)
plt.violinplot(mtcars['mpg'])
```
